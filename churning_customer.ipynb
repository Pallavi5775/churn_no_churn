```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PowerTransformer, StandardScaler, MinMaxScaler, LabelEncoder
from scipy.stats import boxcox



# Step 1: Separate Numeric and Categorical Columns
numeric_cols = df.select_dtypes(include='number')
# Function to detect and encode boolean-like columns dynamically
def detect_and_encode_boolean_columns(df):
    # Define mappings for boolean-like values
    boolean_mappings = {'Yes': 1, 'No': 0, 'True': 1, 'False': 0, 1: 1, 0: 0, 'Missing': None}
    
    # Normalize mappings (case-insensitive)
    normalized_mappings = {str(k).strip().lower(): v for k, v in boolean_mappings.items()}
    
    boolean_cols = []
    
    for col in df.columns:
        # Convert all values to strings, normalize case, and remove NaNs
        unique_values = df[col].dropna().astype(str).str.strip().str.lower().unique()
        
        # Check if all unique values match the normalized boolean keys
        if set(unique_values).issubset(normalized_mappings.keys()):
            boolean_cols.append(col)
            # Map values to 1/0 (or None for 'Missing')
            df[col] = df[col].astype(str).str.strip().str.lower().map(normalized_mappings)
    
    return boolean_cols, df

# Apply the function to the dataset
boolean_cols, df = detect_and_encode_boolean_columns(df)

# Validate results
print(f"Boolean-like columns detected and encoded: {boolean_cols}")

# Function to encode remaining categorical columns
def encode_categorical_columns(df, exclude_cols):
    label_encoders = {}
    encoded_data = pd.DataFrame()
    
    # Select categorical columns excluding boolean columns
    categorical_cols = df.select_dtypes(include=['object']).drop(columns=exclude_cols, errors='ignore')
    
    for col in categorical_cols.columns:
        le = LabelEncoder()
        # Fit and transform the column
        encoded_data[col] = le.fit_transform(categorical_cols[col].astype(str))
        label_encoders[col] = le  # Store the encoder for later use
    
    # Return encoded columns and the encoders
    return encoded_data, label_encoders

# Encode other categorical columns
encoded_categorical_data, label_encoders = encode_categorical_columns(df, boolean_cols)

# Replace original categorical columns with encoded columns
df = pd.concat([df.drop(columns=encoded_categorical_data.columns), encoded_categorical_data], axis=1)


# Step 3: Calculate Skewness for Numeric Columns
skewness = numeric_cols.skew()
print("Skewness of numeric columns:")
print(skewness)

# Step 4: Apply Transformations to Numeric Columns Based on Skewness
transformed_data = {}

for column in numeric_cols.columns:
    if skewness[column] > 0.5:  # Right-skewed
        transformed_data[f"{column}_log"] = np.log1p(numeric_cols[column])
    elif skewness[column] < -0.5:  # Left-skewed
        transformed_data[f"{column}_sqrt"] = np.sqrt(numeric_cols[column])
    else:  # Minimal skew, keep as is
        transformed_data[f"{column}_no_transform"] = numeric_cols[column]

# Convert transformed data to DataFrame
transformed_df = pd.DataFrame(transformed_data)


# Step 6: Combine Transformed Numeric and Encoded Categorical Data
final_data = pd.concat([transformed_df, df], axis=1)

print(final_data)

# Step 7: Scaling the Combined Data (StandardScaler and MinMaxScaler Examples)
scaler_standard = StandardScaler()
scaler_minmax = MinMaxScaler()

scaled_data_standard = scaler_standard.fit_transform(final_data)
scaled_data_minmax = scaler_minmax.fit_transform(final_data)

scaled_df_standard = pd.DataFrame(scaled_data_standard, columns=final_data.columns)
scaled_df_minmax = pd.DataFrame(scaled_data_minmax, columns=final_data.columns)

# Save Transformed DataFrames
transformed_df.to_csv('transformed_data.csv', index=False)
scaled_df_standard.to_csv('scaled_data_standard.csv', index=False)
scaled_df_minmax.to_csv('scaled_data_minmax.csv', index=False)

print("Transformation and scaling complete. Data saved to transformed and scaled CSV files.")

```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import VarianceThreshold


# Load scaled data
scaled_df_standard = pd.read_csv('scaled_data_standard.csv')

# Define the target variable (replace with your actual target column)
target_column = 'Churn'  # Example target column
if target_column not in scaled_df_standard.columns:
    raise ValueError(f"Target column '{target_column}' not found in the dataset.")

# Separate features and target
X = scaled_df_standard.drop(columns=[target_column]).copy()  # Use .copy() to avoid fragmentation
y = scaled_df_standard[target_column]

# Ensure y is 1D
if isinstance(y, pd.DataFrame) or len(y.shape) > 1:
    y = y.values.ravel()

# 1. Feature Generation

# Interaction features
interaction_features = {}
for i, col1 in enumerate(X.columns):
    for j, col2 in enumerate(X.columns):
        if i < j:  # Avoid duplicate and self-interactions
            interaction_features[f"{col1}_x_{col2}"] = X[col1] * X[col2]
            interaction_features[f"{col1}_div_{col2}"] = X[col1] / (X[col2] + 1e-6)  # Avoid division by zero
```

```python
# Assign interaction features to DataFrame
interaction_df = pd.DataFrame(interaction_features)
X = pd.concat([X, interaction_df], axis=1).reset_index(drop=True)

# Polynomial features
polynomial_features = {f"{col}^2": X[col] ** 2 for col in X.columns}
polynomial_df = pd.DataFrame(polynomial_features)
X = pd.concat([X, polynomial_df], axis=1).reset_index(drop=True)

# Log transformations for features with large ranges
log_transformed_features = {f"log_{col}": np.log1p(X[col]) for col in X.columns}
log_df = pd.DataFrame(log_transformed_features)
X = pd.concat([X, log_df], axis=1).reset_index(drop=True)

print("Feature generation complete. Current shape of dataset:", X.shape)
```

```python
from sklearn.feature_selection import VarianceThreshold

# 1.1 Remove low-variance features
selector = VarianceThreshold(threshold=0.01)  # Keep features with variance > 0.01
X_high_variance = selector.fit_transform(X)
selected_features = X.columns[selector.get_support()]
X = pd.DataFrame(X_high_variance, columns=selected_features)

print("Features after removing low-variance ones:", X.shape)

# 1.2 Remove highly correlated features
correlation_matrix = X.corr().abs()
upper_triangle = correlation_matrix.where(
    np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool)
)
to_drop = [column for column in upper_triangle.columns if any(upper_triangle[column] > 0.9)]
X = X.drop(columns=to_drop)

print("Features after removing highly correlated ones:", X.shape)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# Impute missing values
imputer = SimpleImputer(strategy='mean')

# Models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
}

# KFold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Collect metrics for each fold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Define storage for metrics
metrics = {
    "Model": [],
    "Fold": [],
    "Accuracy": [],
    "Precision": [],
    "Recall": [],
    "F1-Score": [],
    "ROC-AUC": []
}

for name, model in models.items():
    for fold, (train_idx, test_idx) in enumerate(kf.split(X)):
        # Split data
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
        
        # Impute missing values in X
        X_train = pd.DataFrame(imputer.fit_transform(X_train), columns=X.columns)
        X_test = pd.DataFrame(imputer.transform(X_test), columns=X.columns)
        
        # Fit model
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_pred_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None
        
        # Compute metrics
        metrics["Model"].append(name)
        metrics["Fold"].append(fold + 1)
        metrics["Accuracy"].append(accuracy_score(y_test, y_pred))
        metrics["Precision"].append(precision_score(y_test, y_pred, zero_division=0))
        metrics["Recall"].append(recall_score(y_test, y_pred, zero_division=0))
        metrics["F1-Score"].append(f1_score(y_test, y_pred, zero_division=0))
        if y_pred_proba is not None:
            metrics["ROC-AUC"].append(roc_auc_score(y_test, y_pred_proba))

# Aggregate results into a DataFrame
import pandas as pd
metrics_df = pd.DataFrame(metrics)
mean_metrics = metrics_df.groupby("Model").mean().reset_index()

# Display results
print("\nAverage Performance Metrics Across Folds:")
print(mean_metrics)
```
